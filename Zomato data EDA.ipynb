```python
df_country=pd.read_excel('Downloads/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
##To check Data types
final_df.dtypes
```

```python
final_df.columns
```

```python
 country_names=final_df.Country.value_counts().index
```

```python
country_value=final_df.Country.value_counts().values
```

```python
##PIE CHART- TOP 3 COUNTRIES
plt.pie(country_value[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation : Zomato maximum records or transactions are from INDIA after that USA and Then United Kingdom 

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

**OBSERVSTION**
1. When Ratings is beytween 4.5 to 4.9--->Excellent
2. When Ratings are between 4.0 to 3.4--->very good
3. When Ratings is between 3.5 to 3.9--->good
4. When Ratings is between 3.0 to 3.4--->average
5. When Ratings is between 2.5 to 2.9--->average
6. When Ratings is between 2.0 to 2.4--->average

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data = ratings,hue="Aggregate rating",palette="rainbow",legend=False)
```

**observation**
1. Not rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
#Count plot 
sns.countplot(x="Rating color", data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','grey'])
```
